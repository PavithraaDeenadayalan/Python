```python
import cv2
import numpy as np

def preprocess_image(image_path, output_path, target_size=(128, 128), to_grayscale=False, display=False):
    """
    Preprocess an image and optionally display it.

    Parameters:
        image_path (str): Path to the input image.
        output_path (str): Path to save the preprocessed image.
        target_size (tuple): Desired size for resizing the image, default is (128, 128).
        to_grayscale (bool): Whether to convert the image to grayscale.
        display (bool): Whether to display the preprocessed image on the screen.
    """
    # Load the image
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"Image at path '{image_path}' could not be loaded.")

    # Resize the image
    resized_image = cv2.resize(image, target_size, interpolation=cv2.INTER_LINEAR)

    # Optionally convert to grayscale
    if to_grayscale:
        preprocessed_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
    else:
        preprocessed_image = resized_image

    # Normalize pixel values to the range [0, 1]
    normalized_image = preprocessed_image / 255.0

    # Convert back to uint8 for saving
    normalized_uint8 = (normalized_image * 255).astype(np.uint8)

    # Save the preprocessed image
    cv2.imwrite(output_path, normalized_uint8)
    print(f"Preprocessed image saved to {output_path}")

    # Display the image if requested
    if display:
        window_name = "Preprocessed Image (Gray)" if to_grayscale else "Preprocessed Image (Color)"
        cv2.imshow(window_name, normalized_uint8)
        cv2.waitKey(0)  # Wait for a key press to close the window
        cv2.destroyAllWindows()

# Example usage
input_image_path = "C:/Users/pavi5/OneDrive/Pictures/wallpaper/image.jpg"  # Path to your input image
output_image_path = "C:/Users/pavi5/OneDrive/Pictures/wallpaper/image2.jpg"  # Path to save the preprocessed image

try:
    preprocess_image(
        input_image_path,
        output_image_path,
        target_size=(128, 128),  # Resize to 128x128 pixels
        to_grayscale=True,       # Set to False to retain color
        display=True             # Set to True to display the preprocessed image
    )
except Exception as e:
    print(f"Error during preprocessing: {e}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def preprocess_image(image_path, output_path, target_size=(128, 128), display=False, color_mode='gray'):
    """
    Preprocesses an image by resizing, converting to grayscale or pseudo-colored, normalizing, saving, 
    and optionally displaying it in a Jupyter Notebook.

    Parameters:
        image_path (str): Path to the input image.
        output_path (str): Path to save the preprocessed image.
        target_size (tuple): Desired size for resizing the image, default is (128, 128).
        display (bool): Whether to display the preprocessed image.
        color_mode (str): 'gray' for grayscale, 'blue' for blue visualization.

    Raises:
        ValueError: If the image cannot be loaded.
        IOError: If the output path is not writable.
    """
    # Step 1: Load the image
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"Image at path '{image_path}' could not be loaded.")

    # Step 2: Resize the image to the specified size
    resized_image = cv2.resize(image, target_size, interpolation=cv2.INTER_LINEAR)

    # Step 3: Convert the image to grayscale
    gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

    # Step 4: Normalize pixel values to the range [0, 1]
    normalized_image = gray_image / 255.0

    # Optional: Convert the normalized image back to uint8 for saving
    normalized_uint8 = (normalized_image * 255).astype(np.uint8)

    # Step 5: Apply pseudo-coloring for blue visualization if requested
    if color_mode == 'blue':
        blue_image = np.zeros((normalized_uint8.shape[0], normalized_uint8.shape[1], 3), dtype=np.uint8)
        blue_image[:, :, 0] = normalized_uint8  # Assign grayscale values to the blue channel
        output_image = blue_image
    else:
        output_image = normalized_uint8  # Grayscale by default

    # Step 6: Save the preprocessed image to the specified output path
    success = cv2.imwrite(output_path, output_image)
    if not success:
        raise IOError(f"Could not write to output path '{output_path}'.")

    print(f"Preprocessed image saved to {output_path}")

    # Step 7: Display the image in Jupyter Notebook (if requested)
    if display:
        plt.figure(figsize=(5, 5))
        if color_mode == 'blue':
            plt.imshow(output_image)  # Display in pseudo-colored blue
        else:
            plt.imshow(output_image, cmap='gray')  # Display in grayscale
        plt.axis('off')
        plt.title(f"Preprocessed Image ({color_mode})")
        plt.show()

# Example usage
input_image_path = "C:/Users/pavi5/OneDrive/Pictures/wallpaper/ogImage.jpg"
output_image_path = "C:/Users/pavi5/OneDrive/Pictures/wallpaper/imageBlue.jpg"

try:
    preprocess_image(input_image_path, output_image_path, target_size=(256, 256), display=True, color_mode='blue')
except Exception as e:
    print(f"Error during preprocessing: {e}")
```

```python
import cv2
import numpy as np
from IPython.display import Video

def preprocess_video(video_path, output_path, target_size=(128, 128)):
    """
    Preprocesses a video by resizing, converting to grayscale, normalizing, and saving it.

    Parameters:
        video_path (str): Path to the input video file.
        output_path (str): Path to save the preprocessed video.
        target_size (tuple): Desired size for resizing each frame, default is (128, 128).

    Raises:
        ValueError: If the video cannot be loaded.
        IOError: If the output path is not writable.
    """
    # Open the input video
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        raise ValueError(f"Video at path '{video_path}' could not be loaded.")

    # Get original video properties
    fps = cap.get(cv2.CAP_PROP_FPS)
    print(f"Original FPS: {fps}")

    # Define codec and create VideoWriter
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Save as MP4
    out = cv2.VideoWriter(output_path, fourcc, fps, target_size, isColor=False)

    print("Processing video...")

    while True:
        ret, frame = cap.read()
        if not ret:  # End of video
            break

        # Resize the frame
        resized_frame = cv2.resize(frame, target_size, interpolation=cv2.INTER_LINEAR)

        # Convert the frame to grayscale
        gray_frame = cv2.cvtColor(resized_frame, cv2.COLOR_BGR2GRAY)

        # Normalize pixel values to range [0, 1]
        normalized_frame = gray_frame / 255.0

        # Convert normalized values back to uint8 for saving
        normalized_uint8 = (normalized_frame * 255).astype(np.uint8)

        # Write the preprocessed frame
        out.write(normalized_uint8)

    print("Video preprocessing complete.")
    
    # Release resources
    cap.release()
    out.release()
    cv2.destroyAllWindows()

    print(f"Preprocessed video saved to {output_path}")

input_video_path = r"D:\video\sri.mp4"
output_video_path = r"D:\video\output.mp4"

try:
    preprocess_video(input_video_path, output_video_path)
    
    # Display the video in Jupyter Notebook
    Video(output_video_path, embed=True)
except Exception as e:
    print(f"Error during preprocessing: {e}")
```

```python
import cv2
import numpy as np
from IPython.display import Video

def preprocess_video(video_path, output_path, target_size=(128, 128)):
    """
    Preprocesses a video by resizing, normalizing, and saving it, preserving color.
    
    Parameters:
        video_path (str): Path to the input video file.
        output_path (str): Path to save the preprocessed video.
        target_size (tuple): Desired size for resizing each frame, default is (128, 128).
    
    Raises:
        ValueError: If the video cannot be loaded.
        IOError: If the output path is not writable.
    """
    # Open the input video
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        raise ValueError(f"Video at path '{video_path}' could not be loaded.")
    
    # Get original video properties
    fps = cap.get(cv2.CAP_PROP_FPS)
    print(f"Original FPS: {fps}")

    # Define codec and create VideoWriter
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Save as MP4
    out = cv2.VideoWriter(output_path, fourcc, fps, target_size, isColor=True)  # Preserve color

    print("Processing video...")

    while True:
        ret, frame = cap.read()
        if not ret:  # End of video
            break
        
        # Resize the frame
        resized_frame = cv2.resize(frame, target_size, interpolation=cv2.INTER_LINEAR)
        
        # Normalize pixel values to range [0, 1]
        normalized_frame = resized_frame / 255.0

        # Convert normalized values back to uint8 for saving
        normalized_uint8 = (normalized_frame * 255).astype(np.uint8)

        # Write the preprocessed frame
        out.write(normalized_uint8)

    print("Video preprocessing complete.")
    
    # Release resources
    cap.release()
    out.release()
    cv2.destroyAllWindows()

    print(f"Preprocessed video saved to {output_path}")

# Example usage
input_video_path = r"D:\video\sri.mp4"
output_video_path = r"D:\video\output.mp4"

try:
    preprocess_video(input_video_path, output_video_path)
    
    # Display the video in Jupyter Notebook
    Video(output_video_path, embed=True)
except Exception as e:
    print(f"Error during preprocessing: {e}")
```
